                                                               Top Winning Teams

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_teams = matches['winner'].value_counts().head(10)

plt.style.use("dark_background")

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_teams.values,
    y=top_teams.index
)

plt.title("Top IPL Teams by Wins", fontsize=16)

plt.xlabel("Wins")
plt.ylabel("Teams")

plt.show()
```

                                                                Top Run Scorers 

```python
top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)

plt.style.use("dark_background")

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_batsmen.values,
    y=top_batsmen.index
)

plt.title("Top IPL Run Scorers", fontsize=16)

plt.xlabel("Runs")
plt.ylabel("Players")

plt.show()
```

                                                                   Top Wicket Takers

```python
top_bowlers = deliveries['bowler'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_bowlers.values,
    y=top_bowlers.index
)

plt.title("Top IPL Wicket Takers", fontsize=16)

plt.xlabel("Wickets")
plt.ylabel("Players")

plt.show()
```

                                                                  Toss Impact Analysis

```python
toss_win = matches[matches['toss_winner'] == matches['winner']]

impact = [
    len(toss_win),
    len(matches) - len(toss_win)
]

labels = [
    'Won Match After Toss',
    'Lost Match After Toss'
]

plt.figure(figsize=(7,7))

plt.pie(
    impact,
    labels=labels,
    autopct='%1.1f%%'
)

plt.title("Impact of Toss on Match Results")

plt.show()
```
